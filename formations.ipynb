```python
import numpy as np
from matplotlib import pyplot as plt
from tqdm import notebook
```

```python
def plot_circle(center, t, radius=1, thickness=0.4):
    plt.fill( center[0] + radius*np.cos(t) , center[1] + radius*np.sin(t), color='#333333')
    plt.fill( center[0] + (radius-thickness)*np.cos(t) , center[1] + (radius-thickness)*np.sin(t), color='#ffffff')
    
```

```python
center1 = (-1,0) #center
center2 = (+1,0) #center
center3 = (0,+np.sqrt(3)) #center
center4 = (0,-np.sqrt(3)) #center

plt.rc('figure', figsize=[8,8])
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
#plt.axis('equal')

t = np.linspace(0, 2*np.pi, 100)
plot_circle(center1, t)
plot_circle(center2, t)
plot_circle(center3, t)
plot_circle(center4, t)

plt.grid(color='lightgray',linestyle='--')
plt.xlim([-3, 3])
plt.ylim([-3, 3])
plt.show()
```

```python
center1 = (-1,-1) #center
center2 = (-1,+1) #center
center3 = (+1,-1) #center
center4 = (+1,+1) #center

plt.rc('figure', figsize=[8,8])
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
#plt.axis('equal')

t = np.linspace(0, 2*np.pi, 100)
plot_circle(center1, t)
plot_circle(center2, t)
plot_circle(center3, t)
plot_circle(center4, t)

plt.grid(color='lightgray',linestyle='--')
plt.xlim([-3, 3])
plt.ylim([-3, 3])
plt.show()
```

```python
def diamond_area(coords, radius=1, thickness=0.4):
    #top circle
    if (coords[0]**2 + (coords[1] - np.sqrt(3))**2) < (radius-thickness)**2:
        return True
    
    #bottom circle
    if (coords[0]**2 + (coords[1] + np.sqrt(3))**2) < (radius-thickness)**2:
        return True
    
    #left circle
    if ((coords[0] + 1)**2 + coords[1]**2) < (radius-thickness)**2:
        return True
    
    #right circle
    if ((coords[0] - 1)**2 + coords[1]**2) < (radius-thickness)**2:
        return True
    
    return False

def box_area(coords, radius=1, thickness=0.4):
    #topleft circle
    if ((coords[0] + 1)**2 + (coords[1] - 1)**2) < (radius-thickness)**2:
        return True
    
    #topright circle
    if ((coords[0] - 1)**2 + (coords[1] - 1)**2) < (radius-thickness)**2:
        return True
    
    #bottomleft circle
    if ((coords[0] + 1)**2 + (coords[1] + 1)**2) < (radius-thickness)**2:
        return True
    
    #bottomright circle
    if ((coords[0] - 1)**2 + (coords[1] + 1)**2) < (radius-thickness)**2:
        return True
    
    return False

def count_in(points, funct):
    total = len(points)
    cup = 0
    for e in points:
        if funct(e):
            cup += 1
    return cup / total
```

```python
mean = [0,0]
cov = np.array([[0.3,0], [0,0.92]])
a = np.random.multivariate_normal(mean, cov, 10000)
print(count_in(a, diamond_area))


center1 = (-1,0) #center
center2 = (+1,0) #center
center3 = (0,+np.sqrt(3)) #center
center4 = (0,-np.sqrt(3)) #center

plt.rc('figure', figsize=[8,8])
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
#plt.axis('equal')

t = np.linspace(0, 2*np.pi, 100)
plot_circle(center1, t)
plot_circle(center2, t)
plot_circle(center3, t)
plot_circle(center4, t)

for e in notebook.tqdm(a):
    if diamond_area(e):
        col = "#44ff4488"
    else:
        col = "#ff444488"
    plt.scatter(e[0], e[1], s=14, color=col)

plt.grid(color='lightgray',linestyle='--')
plt.xlim([-3, 3])
plt.ylim([-3, 3])
plt.show()
```

```python
# Diamond aiming middle

mean = [0,0]
size = 101
num_gauss_iterations = 6000
desvx = desvy = np.linspace(0, 2, size)
z = []
for dx in notebook.tqdm(desvx):
    for dy in desvy:
        cov = np.array([[dx,0], [0,dy]])
        a = np.random.multivariate_normal(mean, cov, num_gauss_iterations)
        z.append(count_in(a, diamond_area))

Z = np.asarray(z).reshape(size, size).transpose()
plt.rc('figure', figsize=[9,7.25])
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
plt.contourf(desvx, desvy, Z, levels=np.linspace(0,1,101), cmap = 'magma')
plt.colorbar(ticks=np.linspace(0,1,11))
plt.axis('equal')
plt.xlim([0, 2])
plt.ylim([0, 2])

plt.title(r'Probability of scoring in diamond aiming middle', fontsize=20)
plt.xlabel(r'Deviation in x-axis $\sigma_x$', fontsize=16)
plt.ylabel(r'Deviation in y-axis $\sigma_y$', fontsize=16)

plt.show()

ind = np.unravel_index(np.argmax(Z, axis=None), Z.shape)
print('Diamond aiming middle')
print('Max prob: ', np.max(Z), ' in desv: (', round(desvx[ind[0]], 2), ",", round(desvy[ind[1]], 2), ")")
print('Avg prob: ', np.average(Z))
```
